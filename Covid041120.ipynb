```
import altair as alt
import pandas as pd
import numpy as np
```

```
covid_Nov=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/11-02-2020.csv')
```

```
map = alt.topo_feature("https://cdn.jsdelivr.net/npm/world-atlas@2/countries-110m.json", "countries")
```

```
background = alt.Chart(map).mark_geoshape(
    fill='grey',
    stroke='white'
).properties(
    width=800,
    height=400
).project('naturalEarth1')

points = alt.Chart(covid_Nov.loc[covid_Nov['Confirmed'] > 0]).mark_circle().encode(
    longitude='Long_:Q',
    latitude='Lat:Q',
    tooltip=['Country_Region:N','Province_State:N','Confirmed:Q', 'Deaths:Q','Recovered :Q'],
    color=alt.value('#5d00ff'),
).properties(
    title='Covid 19 confirmed cases with death and recovery numbers per region'
)

background + points
```

```

alt.Chart(covid_Nov).mark_circle().encode(
    x='Deaths:Q',
    y='Confirmed:Q',
    color='Country_Region:N',
    tooltip=[alt.Tooltip('Country_Region:N'),
             alt.Tooltip('Confirmed:Q'),
             alt.Tooltip('Deaths:Q')]
             
).interactive()
```
